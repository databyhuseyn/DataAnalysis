```python
plt.figure(figsize=(20,10),facecolor='white')
sns.violinplot(x='EmpDepartment',y='PerformanceRating',hue=data.Gender,data=data,palette='rocket')
plt.xlabel('EmpDepartment',fontsize=20)
plt.ylabel('PerformanceRating',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(20,10),facecolor='white')
ax = sns.countplot(x='EmpDepartment',hue=data.PerformanceRating,data=data)
for i in ax.patches:
    ax.annotate(f'{i.get_height()}', (i.get_x()+0.06, i.get_height()+2))
plt.title('Employee Departments vs PerformanceRating',fontsize=20)
plt.xlabel('EmpDepartment',fontsize=20)
plt.ylabel('PerformanceRating',fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.barplot(x='Age', y='PerformanceRating', data=data)
plt.title('Age vs. Performance Rating')
plt.show()
```

```python
import numpy as np

numerical_cols = ['Age', 'EmpHourlyRate', 'EmpEnvironmentSatisfaction', 'TotalWorkExperienceInYears', 'ExperienceYearsAtThisCompany', 'PerformanceRating']
numerical_data = data[numerical_cols]

plt.figure(figsize=(12, 8))
corr_matrix = numerical_data.corr()

plt.imshow(corr_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar()

plt.xticks(np.arange(len(numerical_cols)), numerical_cols, rotation=45)
plt.yticks(np.arange(len(numerical_cols)), numerical_cols)

# Add annotations
for i in range(len(numerical_cols)):
    for j in range(len(numerical_cols)):
        plt.text(j, i, f'{corr_matrix.iloc[i, j]:.2f}', ha='center', va='center', color='w')

plt.title('Correlation Heatmap')
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming numerical_cols is a list of numerical columns in your dataset
numerical_cols = ['Age', 'EmpHourlyRate', 'EmpEnvironmentSatisfaction', 'TotalWorkExperienceInYears', 'ExperienceYearsAtThisCompany', 'PerformanceRating']
numerical_data = data[numerical_cols]

# Add more advanced relationships using pair plot
sns.pairplot(data, vars=numerical_cols, hue='PerformanceRating', palette='viridis', diag_kind='kde')
plt.suptitle('Pair Plot of Numerical Variables', y=1.02)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'Gender', 'EmpDepartment', and 'PerformanceRating' are columns in your dataset
gender_department_performance = data.groupby(['Gender', 'EmpDepartment'])['PerformanceRating'].mean().reset_index()

# Plotting with Seaborn
plt.figure(figsize=(12, 8))
sns.barplot(x='EmpDepartment', y='PerformanceRating', hue='Gender', data=gender_department_performance, palette='pastel')
plt.title('Performance Rating by Gender and Department')
plt.xlabel('Department')
plt.ylabel('Mean Performance Rating')
plt.legend(title='Gender')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# 1. Correlation Analysis
correlation_matrix = data.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix[['PerformanceRating']], annot=True, cmap='coolwarm')
plt.title('Correlation Matrix with PerformanceRating')
plt.show()
```

```python
# 2. Distribution and Descriptive Statistics
plt.figure(figsize=(10, 6))
sns.histplot(data['PerformanceRating'], kde=True)
plt.title('Distribution of PerformanceRating')
plt.show()
```

```python
categorical_features = ['Gender', 'EducationBackground', 'MaritalStatus', 'EmpDepartment', 'EmpJobRole', 'BusinessTravelFrequency', 'OverTime']
for feature in categorical_features:
    plt.figure(figsize=(12, 6))
    sns.boxplot(x=feature, y='PerformanceRating', data=data)
    plt.title(f'{feature} vs PerformanceRating')
    plt.show()
```

```python
# 4. Age and Experience
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Age', y='PerformanceRating', data=data)
plt.title('Age vs PerformanceRating')
plt.show()

```

```python
# 5. Work Environment and Satisfaction
satisfaction_features = ['EmpEnvironmentSatisfaction', 'EmpJobSatisfaction', 'EmpRelationshipSatisfaction', 'EmpWorkLifeBalance']
for feature in satisfaction_features:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=feature, y='PerformanceRating', data=data)
    plt.title(f'{feature} vs PerformanceRating')
    plt.show()
```

```python
# 6. Training and Development
plt.figure(figsize=(10, 6))
sns.boxplot(x='TrainingTimesLastYear', y='PerformanceRating', data=data)
plt.title('TrainingTimesLastYear vs PerformanceRating')
plt.show()
```

```python
# 7. Attrition and Overtime
attrition_plot = sns.catplot(x='Attrition', y='PerformanceRating', kind='box', data=data)
overtime_plot = sns.catplot(x='OverTime', y='PerformanceRating', kind='box', data=data)
```
